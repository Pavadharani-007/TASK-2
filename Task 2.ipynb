```python
# -----------------------------
# IMPORTING LIBRARIES
# -----------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display all columns
pd.set_option('display.max_columns', None)
```

```python
# -----------------------------
# LOAD DATASET
# -----------------------------
df = pd.read_csv(r"C:\Users\HP\Downloads\IRIS - Copy.csv")
df.head()
```

```python
# -----------------------------
# 2. EXPLORE DATA STRUCTURE
# -----------------------------
print("ðŸ“Œ FIRST FIVE ROWS")
display(df.head())

print("\nðŸ“Œ DATA INFO")
print(df.info())

print("\nðŸ“Œ SUMMARY STATISTICS")
display(df.describe(include='all'))

print("\nðŸ“Œ DATA TYPES")
print(df.dtypes)

print("\nðŸ“Œ SHAPE OF DATA")
print(df.shape)
```

```python
# -----------------------------
# 3. CHECK MISSING VALUES
# -----------------------------
print("ðŸ“Œ MISSING VALUES")
print(df.isnull().sum())

plt.figure(figsize=(8,5))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Value Heatmap")
plt.show()
```

```python
# -----------------------------
# 4. IDENTIFY TRENDS & PATTERNS
# -----------------------------
df.hist(figsize=(12, 6))
plt.suptitle("Distribution of Numeric Columns")
plt.show()
```

```python
sns.pairplot(df.select_dtypes(include=['float64', 'int64']))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# -----------------------------
# 5. DETECT OUTLIERS
# -----------------------------
numeric_cols = df.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    plt.figure(figsize=(6, 3))
    sns.boxplot(x=df[col])
    plt.title(f"Outlier Detection â€” {col}")
    plt.show()
```

```python
# -----------------------------
# 6. TEST SIMPLE HYPOTHESES
# -----------------------------
if 'Category' in df.columns and 'Price' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df, x='Category', y='Price')
    plt.title("Price Variation Across Categories")
    plt.xticks(rotation=45)
    plt.show()

if 'Price' in df.columns and 'Rating' in df.columns:
    sns.scatterplot(data=df, x='Rating', y='Price')
    plt.title("Rating vs Price")
    plt.show()
```

```python
# -----------------------------
# 7. DETECT DATA ISSUES
# -----------------------------
print("ðŸ“Œ DUPLICATES:", df.duplicated().sum())

numeric_cols = df.select_dtypes(include=[np.number]).columns
print("\nðŸ“Œ ZERO/NEGATIVE VALUES CHECK")
print((df[numeric_cols] <= 0).sum())
```
